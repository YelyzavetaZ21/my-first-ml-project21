```python
import pandas as pd
import plotly.express as px

data = {
    'Country': [
        'Malta', 'Latvia', 'Estonia', 'Hungary', 'Croatia', 'Slovakia',
        'United Kingdom', 'Romania', 'Greece', 'Poland',
        'Germany', 'France', 'Spain', 'Italy', 'Netherlands'
    ],
    'Obesity_Percent_2022_2023': [25.7, 23.9, 22.9, 22.0, 21.5, 20.8, 27.8, 19.5, 18.0, 17.5, 16.5, 15.8, 14.2, 11.7, 13.5],
    'Region': ['South', 'Baltic', 'Baltic', 'Central', 'South', 'Central', 'North-West', 'East', 'South', 'East', 'Central', 'West', 'South', 'South', 'West']
}

df = pd.DataFrame(data)

df['Obesity_Percent_2025_estimate'] = df['Obesity_Percent_2022_2023'] + 1.2

print("Top 5 countries by obesity rate ~2022–2023:")
print(df.sort_values('Obesity_Percent_2022_2023', ascending=False).head())

print("\nEstimated obesity rates for 2025:")
print(df[['Country', 'Obesity_Percent_2025_estimate']]
      .sort_values('Obesity_Percent_2025_estimate', ascending=False)
      .head())
```

```python
import plotly.express as px

fig = px.choropleth(
    df,
    locations='Country',
    locationmode='country names',
    color='Obesity_Percent_2022_2023',
    hover_name='Country',
    hover_data={'Obesity_Percent_2022_2023': ':.1f'},
    color_continuous_scale='OrRd',
    title='Adult Obesity Prevalence (BMI ≥ 30) in Selected European Countries (~2022–2023)',
    scope='europe',
    height=600
)

fig.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=True,
        showcountries=True,
        countrycolor='gray',
        resolution=50
    ),
    title_x=0.5,
    margin=dict(l=0, r=0, t=60, b=0)
)

fig.update_traces(
    marker_line_width=0.5,
    marker_line_color='darkgray'
)

fig.show()
```

```python
df_top = df.sort_values('Obesity_Percent_2022_2023', ascending=False).head(10)

fig_bar = px.bar(
    df_top,
    x='Country',
    y=['Obesity_Percent_2022_2023', 'Obesity_Percent_2025_estimate'],
    barmode='group',
    title='Top 10 European Countries by Adult Obesity Rate + 2025 Estimate',
    labels={
        'value': 'Obesity Rate (%)',
        'variable': 'Period',
        'Country': 'Country'
    },
    height=500,
    color_discrete_sequence=['#ff7f0e', '#d62728']
)

fig_bar.update_layout(
    xaxis_title='Country',
    yaxis_title='Obesity Rate (%)',
    legend_title='Period',
    barmode='group',
    bargap=0.15,
    bargroupgap=0.1,
    title_x=0.5,
    xaxis_tickangle=-45
)

fig_bar.show()
```
